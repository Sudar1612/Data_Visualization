```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[3,6,9,12,15]
plt.xlabel("X-Axis",fontsize=10)
plt.ylabel("Y1-Axis",fontsize=10)
plt.ylabel("Y2-Axis",fontsize=10)
plt.title("Line Plot",fontsize=15)
plt.plot(x,y1,color="blue",label="Graph 1")
plt.plot(x,y2,color="black",label="Graph 2")
plt.legend(loc=2)
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.xlabel("X-Axis",fontsize=10)
plt.ylabel("Y1-Axis",fontsize=10)
plt.title("Bar Plot",fontsize=15)
plt.bar(x,y,color="black",width=0.75)
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[3,6,9,12,15]
plt.xlabel("X-Axis",fontsize=10)
plt.ylabel("Y1-Axis",fontsize=10)
plt.ylabel("Y2-Axis",fontsize=10)
plt.title("Scatter Plot",fontsize=15)
plt.scatter(x,y1,color="blue",label="Graph 1")
plt.scatter(x,y2,color="red",label="Graph 2")
plt.legend(loc=2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
import numpy as np
sample=np.random.randint(500,1000)
plt.hist(sample,color="black")
plt.show()
```

```python
plt.figure(figsize=(10,10))
slices=[10,20,30,40,50]
act=["A","B","C","D","E"]
cols=["black","green","purple","red","blue"]
plt.pie(slices,labels=act,colors=cols,autopct="%1.2f",explode=(0.005,0.005,0.005,0.005,0.005))
plt.show() 
```
